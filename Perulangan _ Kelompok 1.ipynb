```
target = 25
for x in range(target + 1):
  for y in range(target - x + 1):
    z = target - x - y
    print(f"x = {x}, y = {y}, z = {z}")
```

```
def bank_simulation():
    saldo = 100000  # Saldo awal
    saldo_minimum = 10000

    while True:
        print(f"Saldo saat ini: Rp{saldo}")
        kode = int(input("Masukkan kode transaksi (0: Setor, 1: Tarik, 2: Keluar): "))

        if kode == 2:
            print("Terima kasih telah menggunakan layanan kami.")
            break

        jumlah = int(input("Masukkan jumlah uang: "))

        if kode == 0:
            saldo += jumlah
            print(f"Anda telah menyetor Rp{jumlah}. Saldo baru: Rp{saldo}")
        elif kode == 1:
            if saldo - jumlah >= saldo_minimum:
                saldo -= jumlah
                print(f"Anda telah menarik Rp{jumlah}. Saldo baru: Rp{saldo}")
            else:
                print("Transaksi gagal! Saldo tidak boleh kurang dari Rp10.000.")
        else:
            print("Kode transaksi tidak valid.")

bank_simulation()
```
